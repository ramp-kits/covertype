```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import StratifiedShuffleSplit
pd.set_option('display.max_columns', 500)
```

```python
data = datasets.fetch_openml('covertype', version=3)
```

```python
data_df = pd.DataFrame(data.data, columns=data.feature_names)
data_df['Cover_Type'] = data.target
data_df.head()
```

```python
data_df.dtypes
```

```python
data_df.describe()
```

```python
cv = StratifiedShuffleSplit(n_splits=1, test_size=0.5, random_state=57)
```

```python
train_is, test_is = list(cv.split(data_df, data_df['Cover_Type']))[0]
```

```python
test_is
```

```python
data_df.iloc[train_is].to_csv('data/train.csv.gz', compression='gzip', index=False)
```

```python
data_df.iloc[test_is].to_csv('data/test.csv.gz', compression='gzip', index=False)
```
